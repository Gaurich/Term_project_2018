# AAPL

Selecting the adjusted monthend close price for AAPL 

```python
data_aapl = quandl.get_table('WIKI/PRICES', ticker = ['AAPL'], 
                        qopts = { 'columns': ['ticker', 'date', 'adj_close'] }, 
                        date = { 'gte': '2013-2-28', 'lte': '2018-2-28' }, 
                        paginate=True)
data_aapl.head()
```

```python
data_aapl = quandl.get_table('WIKI/PRICES', ticker = ['AAPL'], 
                        qopts = { 'columns': ['adj_close'] }, 
                        date = { 'gte': '2013-2-28', 'lte': '2018-2-28' }, 
                        paginate=True)

# Checking if the month end close prices are auto-correlated with time lag

autocorrelation_plot(data_aapl)
```

Intrepetation of above plot:

The autocorrelation with lag zero always equals 1, because this represents the autocorrelation between each term and itself. Price and price with lag zero are the same variable.
As per the definition of autocorrelation, we know that - If a spike is significantly different from zero, that is evidence of autocorrelation. A spike that’s close to zero is evidence against autocorrelation.
 
From value 0 to apprx 300 the month end prices are correlated with time lags i.e. when the price of Apple stock rises, it tends to continue rising. When the price of Apple stock falls, it tends to continue falling. This figure illustrates the same.
Similarly between 700 and 1200 the prices are correlated . 


```python
#Calculating returns
close_aapl = data_aapl['adj_close']
rets_aapl = close_aapl / close_aapl.shift(1) - 1
rets_aapl.head()
```

```python
#rets_aapl = close_aapl / close_aapl.shift(1) - 1
#rets_aapl.head()
```

```python
#drop NaN to plot autocorrelation
rets_aapl = rets_aapl.dropna()
```

```python
#rets_aapl.head()
```

```python
len(rets_aapl)
```

```python
autocorrelation_plot(rets_aapl)
```

Intrepetation of above plot:
    
The autocorrelation plot for daily returns to AAPL stock shows that the spikes are not statistically 
significant. This indicates that the returns are not  correlated, as shown here.

The values tend to fluctuate randomly around zero. 
This means that the returns are largely independent of each other.
